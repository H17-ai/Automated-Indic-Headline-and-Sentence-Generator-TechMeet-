```python
import re
import os
import keras 
import gensim
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pprint import pprint
%matplotlib inline
def process_text(word):
    char_set=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    alpToDig=dict()
    digToAlp=dict()
    for i,j in enumerate(char_set):
      alpToDig[j]=i+1
      digToAlp[i+1]=j   
    processed_frame_x=np.zeros((len(word),29,len(char_set)),dtype='int8')
    for sample_index,sample in enumerate(word):
        for char_index, char in enumerate(sample.lower()):
            processed_frame_x[sample_index, char_index-1, alpToDig[char]-1]=1
    return processed_frame_x
```

```python
from keras.models import load_model
model = load_model('/content/drive/MyDrive/Automatic Headline generator/_assets/hinglish_classifier_1.h5')
```

```python
words = [i for i in gensim.summarization.textcleaner.tokenize_by_word()]
processed_words = process_text(words)
pred = model.predict(processed_words)
for i,p in enumerate(pred):
    if(pred[i].round()==0):
        print('%s is English'%words[i])
    else:
        print('%s is Hinglish'%words[i])
```
